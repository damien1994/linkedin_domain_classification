```python
# lowercase all job title
df['Titre'] = df.Titre.str.lower()
```

```python
# remove accents
def remove_accents(a):
    return unidecode.unidecode(str(a))

df['Titre'] = df['Titre'].apply(remove_accents)
```

```python
from bs4 import BeautifulSoup
import urllib.request
import csv
```

```python
# specify the url
urlpage = 'https://coders.bluecoders.com/jobs'
```

```python
# query the website and return the html to the variable 'page'
page = urllib.request.urlopen(urlpage)
# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup)
```

```python
# find results within table
table = soup.find('table', attrs={'class': 'class="jobs col-xs-12"'})
```
